```python
import pandas as pd 
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer  
import numpy as np
import re
```

```python
def invalidity_clear(text):
    text = re.sub(r"https?:\/\/(www\.)?[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)", '', text)
    text = re.sub(r'[^\w\s]', '', text, re.UNICODE)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r"<.*?>", ' ', text)
    text = re.sub(r"[^\u0000-\u007F]+?", " ", text)
    text = text.lower()
    text = text.strip()
    text 
    return text
```

```python
def lemmatizer(text):
      stop_words=set(nltk.corpus.stopwords.words('english'))
      word_tokenize=nltk.tokenize.word_tokenize
      le=WordNetLemmatizer()
      word_tokens=word_tokenize(text)
      tokens=[le.lemmatize(w) for w in word_tokens if w not in stop_words and len(w)>3]
      cleaned_text=" ".join(tokens)
      return cleaned_text
```

```python
df1 = pd.read_csv('../data/BA_reviews.csv.zip', usecols=['review'])
df2 = pd.read_csv('../data/amazon_reviews.csv', usecols=['reviews.text'])
df3 = pd.read_csv('../data/McDonald_s_Reviews.csv.zip', encoding='latin-1', usecols=['review'])
```

```python
df1['review'] = df1['review'].apply(invalidity_clear)
df1['review'] = df1['review'].apply(lemmatizer)
df1
```

```python
df2['review'] = df2['reviews.text'].apply(invalidity_clear)
df2['review'] = df2['reviews.text'].apply(lemmatizer)
df2.drop(['reviews.text'], axis=1, inplace=True)
```

```python
df3['review'] = df3['review'].apply(invalidity_clear)
df3['review'] = df3['review'].apply(lemmatizer)
df3
```

```python
df_main = pd.concat([df1, df2, df3])
df_main
```

```python
df_main['review'] = df_main['review'].str.replace('flight', '')
df_main['review'] = df_main['review'].str.replace('airline', '')
df_main['review'] = df_main['review'].str.replace('plane', '')
df_main['review'] = df_main['review'].str.replace('airport', '')
df_main['review'] = df_main['review'].str.replace('mcdonald', '')
df_main['review'] = df_main['review'].str.replace('mcdonalds', '')
df_main['review'] = df_main['review'].str.replace('british', '')
df_main['review'] = df_main['review'].str.replace('great', '')
df_main['review'] = df_main['review'].str.replace('good', '')
df_main['review'] = df_main['review'].str.replace('bad', '')
df_main['review'] = df_main['review'].str.replace('love', '')
df_main['review'] = df_main['review'].str.replace('like', '')
df_main['review'] = df_main['review'].str.replace(' s ', '')
```

```python
df_main.dropna(inplace=True)
```

```python
df_main.to_csv('final_dataset_clean.csv', index=False)
```
